### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
#print(soup)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find_all("em")[-1].text,'\t', d.find_all('a')[-1].text)
   
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#34個類別
count=[0 for i in range(0, 34)]
index=[0 for i in range(0, 34)]

for j in range(0, 34):
    index[j]=[]

category=['財經','論壇','國際','大陸','社會','地方','新奇','生活','寵物動物','影劇','體育','旅遊','消費','名家','ET來了',\
          '3C家電','健康','男女','公益','遊戲','電影','時尚','網搜','電商','親子','房產雲','ET車雲','軍武','保險','法律',\
         '直銷雲','探索','電競','政治']
#index裡面是content

for d in soup.find(class_="part_list_2").find_all('h3'):
    for j in range(0, 34):
        if d.find_all("em")[-1].text==category[j]:
            count[j]+=1
            index[j]+=[d.find(class_="date").text, d.find_all('a')[-1].text]

#將儲存的title印出來            
for order in range(0, 34):
    print('***',category[order],'*** : ')

    if index[order]!=[]:
        try:
            for i in range(0,max(count)):
                print(index[order][i])
        except:
            print('')
              
    else:
        print('沒有相關新聞QQ\n')
    
            
        


   
```

### ③ 哪一個類別的文章最多

```python
for i in range(0, 34):
    if count[i]==max(count):
        print(category[i],'類的文章最多，共有',max(count),'篇!!')
    
```
