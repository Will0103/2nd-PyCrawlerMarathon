## 作業目標

## 1. 取出今天所有的新聞

```python
#今天3/20
browser.quit();
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
import time
for i in range(1000):
    time.sleep(0.01)
    browser.execute_script("window.scrollTo(0, 1000000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[5:10]=='03/20':
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
browser.quit();
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
import time
for i in range(100):
    time.sleep(0.01)
    browser.execute_script("window.scrollTo(0, 100000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
now_hr=int(soup.find(class_="date").text[11:13])
now_min=int(soup.find(class_="date").text[14:16])

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[11:13]==str(now_hr-1) :
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    elif d.find(class_="date").text[11:13]==str(now_hr-2):
        if int(d.find(class_="date").text[14:16])>=now_min:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
    elif d.find(class_="date").text[11:13]==str(now_hr):
        if int(d.find(class_="date").text[14:16])<=now_min:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
#3天前 0317
browser.quit();
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select


browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value('17')
browser.find_element_by_id('button').click()
import time
for i in range(1000):
    time.sleep(0.01)
    browser.execute_script("window.scrollTo(0, 1000000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[5:13]=='03/17 15' or d.find(class_="date").text[5:13]=='03/17 16':
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
