## requests + BeautifulSoup


```python
#HW8知乎網站
import requests
url = 'https://www.zhihu.com/explore'

r = requests.get(url,headers=headers)
response = r.text
r.encoding = 'utf-8'
from bs4 import BeautifulSoup
soup = BeautifulSoup(response[0:5000], "html5lib")
print(soup)
```

## Grab + PyQuery

```python
url = 'https://www.zhihu.com/explore'
from grab import Grab
g = Grab()
resp = g.go(url)
from pyquery import PyQuery as pq
doc = pq(resp.body)
print(doc)
```
